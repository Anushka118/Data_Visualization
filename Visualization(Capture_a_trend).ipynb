```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/drive/MyDrive/data/num_trend.xlsx')
df
```

```python
df_cnt = df[df['stat']=='cnt']
df_nmis = df[df['stat']=='nmis']
df_avg = df[df['stat']=='avg']
df_min = df[df['stat']=='min']
df_max = df[df['stat']=='max']

print(df_cnt.shape)
print(df_nmis.shape)
print(df_avg.shape)
print(df_min.shape)
print(df_max.shape)
```

```python
df_cnt.drop(columns=['stat'], inplace = True)
df_nmis.drop(columns=['stat'], inplace = True)
df_avg.drop(columns=['stat'], inplace = True)
df_min.drop(columns=['stat'], inplace = True)
df_max.drop(columns=['stat'], inplace = True)
```

```python
df_cnt
```

```python
df_cnt_pivot = df_cnt.pivot(index = 'field', columns = 'yymm', values = 'value')
df_cnt_pivot.index.name = None

df_avg_pivot = df_avg.pivot(index = 'field', columns = 'yymm', values = 'value')
df_avg_pivot.index.name = None

df_nmis_pivot = df_nmis.pivot(index = 'field', columns = 'yymm', values = 'value')
df_nmis_pivot.index.name = None

df_min_pivot = df_min.pivot(index = 'field', columns = 'yymm', values = 'value')
df_min_pivot.index.name = None

df_max_pivot = df_max.pivot(index = 'field', columns = 'yymm', values = 'value')
df_max_pivot.index.name = None
```

```python
df_cnt_pivot.columns = df_cnt_pivot.columns.astype(str)
df_avg_pivot.columns = df_avg_pivot.columns.astype(str)
df_nmis_pivot.columns = df_nmis_pivot.columns.astype(str)
df_min_pivot.columns = df_min_pivot.columns.astype(str)
df_max_pivot.columns = df_max_pivot.columns.astype(str)
```

```python
# create a dropdown widget to select the field
stats_df = {
    'cnt': df_cnt_pivot,
    'nmis': df_nmis_pivot,
    'avg': df_avg_pivot,
    'min': df_min_pivot,
    'max': df_max_pivot
}
field_options = list(df['field'].unique())
df_dropdown = widgets.Dropdown(
    options=list(stats_df.keys()),
    description='Stat:'
)
field_dropdown = widgets.Dropdown(
    options=field_options,
    description='Field:'
)

#function to update the field dropdown based on the selected dataframe
def update_field_dropdown(df_name):
  df = stats_df[df_name]
  field_dropdown.options = df.index.unique()

#function to update the plot based on the selected field

def update_plot(df_name, field):
        df = stats_df[df_name]
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), gridspec_kw={'height_ratios': [3, 1]}, dpi=300) # Adjust height_ratios as needed

        # Line plot in the first subplot
        palette = sns.color_palette(palette='GnBu')
        line_color = palette[3]
        sns.set_style("whitegrid")
        sns.lineplot(x=df.columns, y=df.loc[field], marker='o', linewidth=0.75, markersize=5, color=line_color, ax=ax1)
        ax1.invert_xaxis()  # Reverse x-axis
        ax1.set_xticks(df.columns)
        ax1.set_xticklabels(df.columns, rotation=45, ha='right')
        ax1.set_xlabel('YYMM', fontsize=8)
        ax1.set_ylabel(f'{df_name}', fontsize=8)
        ax1.set_title(f'{field} ({df_name})', fontsize=10)
        sns.despine()


        # Table in the second subplot
        table_data = df.loc[[field]].T.reset_index()  # Transpose and reset index for table format
        table_data.columns = ['YYMM', df_name]  # Rename columns
        table = ax2.table(cellText=table_data.values, colLabels=table_data.columns, loc='center', cellLoc='center')
        table.auto_set_font_size(False)
        table.set_fontsize(8)
        table.scale(1, 1.5)  # Adjust scale as needed
        ax2.axis('off')  # Hide axes for the table subplot


        plt.tight_layout()
        plt.show()

widgets.interactive(update_field_dropdown, df_name = df_dropdown)
widgets.interactive(update_plot,df_name = df_dropdown, field = field_dropdown)
```
